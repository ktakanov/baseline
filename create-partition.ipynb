```python
import pandas as pd

df_path = 'data/yoochoose-clicks.dat' 
df = pd.read_csv(df_path,sep=',',header=None,names=['session_id','timestamp','item_id','category'])
df.head()
```

```python
sids = df.session_id.values
p_25 = sids[: int(len(sids) * 0.25)]
p_25
```

```python
df = df[df.session_id.isin(p_25)]
df.head()
```

```python
df.to_csv("data/yoochoose-clicks-025.dat")
```
